### Matrix Multiplication

Use the `@` operator to do matrix multiplication with NumPy arrays for M :

Let's compute $2I + 3A - AB$ for

$$
A = \begin{bmatrix}
1 & 3 \\\
-1 & 7
\end{bmatrix}
\ \ \ \
B = \begin{bmatrix}
5 & 2 \\\
1 & 2
\end{bmatrix}
$$

and $I$ is the identity matrix of size 2:

```python
2*I + 3*A - A@B
```
